# Generating the Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
cols = ["workclass","marital-status","occupation","relationship","race","sex","native-country","age","fnlwgt","education-num","capital-gain","capital-loss","hours-per-week","class"]
```

```python
class Node():
    def __init__(self, feature_index=None, threshold=None, children=None, node_dataset =None,IG=None, value=None):
        ''' constructor ''' 
        
        # for decision node
        self.feature_index = feature_index
        self.threshold = threshold
        self.children=children
        self.node_dataset=node_dataset
        self.IG = IG
        
        # for leaf node
        self.value = value
```

```python
class DecisionTreeClassifier():
    
    def __init__(self, min_samples_split=2, maxDepth=2):
        self.root = None
        # stopping conditions
        self.min_samples_split = min_samples_split
        self.maxDepth = maxDepth
    
    def buildTree(self,data,currentDepth=0):        

        sampleSpace = data.iloc[:,:-1]
        result = data.iloc[:,-1]
        sampleSize , attributes = sampleSpace.shape
        if currentDepth <= self.maxDepth and sampleSize>=self.min_samples_split:
            best_split = self.get_split(data,sampleSize,attributes)
            if best_split["IG"] > 0:
                children=[]
                for idx in range(0,len(best_split["children"])):
                    children.append(self.buildTree(best_split["children"][idx], currentDepth+1))
                return Node(best_split["feature-index"], best_split["threshold"], children, data, best_split["IG"])
        leaf_value = self.getLeafValue(result)
        return Node(value = leaf_value)
    
    def getLeafValue(self, y):
        #convert to array
        if len(y[y==1])>=len(y[y==0]):
            return 1
        return 0
 
    def IGain(self, parent, children, mode="entropy"):
        weight=[]
        for child in children:
            weight.append(len(child)/len(parent))
        if mode=="gini":
            gain = self.gini_index(parent)
            idx=0
            for child in children:
                gain -= weight[idx]*self.gini_index(child)
                idx=idx+1
            return gain
        
        gain = self.entropy(parent)
        idx=0
        for child in children:
            gain -= weight[idx]*self.entropy(child)
            idx=idx+1
        return gain
    
    def get_split(self,data,sampleSize,attributes):
        
        best_split = {}
        IGmax = -float("inf")
        for attribute in range(attributes):
            
            attribute_values = data.iloc[:, attribute]
            possible_values=np.unique(attribute_values)
            idx=0
            dataArr=[]
            infoArr=[]
            thresholdArr=[]
            for value in possible_values:
                temp_attr_dataset = data.iloc[:,:]
                temp_result_dataset = data.iloc[:,-1]
                child_data = self.dataset_split(temp_attr_dataset, attribute, value)
                dataArr.append(child_data)
                infoArr.append(child_data.iloc[:,-1])
                thresholdArr.append(possible_values[idx])
                idx=idx+1
            IG = self.IGain(data.iloc[:,-1],infoArr,"gini")
            if IG>IGmax:
                best_split["children"]=dataArr
                best_split["feature-index"]=attribute
                best_split["threshold"]=thresholdArr
                best_split["IG"]=IG
                IGmax=IG
        return best_split
            
                
    def dataset_split(self,data,attribute,value):
        return (data[data.iloc[:, attribute] == value])
    
    def entropy(self, y):
        values=[0,1]

        entropy = 0
        for value in values:
            p_cls = len(y[y == value]) / len(y)
            if(p_cls>0):
                entropy += -p_cls * np.log2(p_cls)
        return entropy
                         
    def gini_index(self, y):
        values=[1,2]
        gini = 0
        for cls in values:
            p_cls = len(y[y == cls]) / len(y)
            gini += p_cls**2
        return 1 - gini
                         
    def fit(self, data):
        self.root = self.buildTree(data)
                         
    def print_tree(self, tree=None, indent=""):
        
        if not tree:
            tree = self.root
        if tree.value is not None:
            print(indent,tree.value)
        else:
            idx=0
            for child in tree.children:
                print(indent + str(cols[tree.feature_index]), "==", tree.threshold[idx], "?")
                idx=idx+1
                self.print_tree(child, indent + "  ")
    
    def predict(self, X,tree=None):
        if tree is None:
            tree = self.root
        rows,cols=X.shape
        predictions=[]
        idx=0
        for row in range(rows):
            predictions.append(self.make_prediction(X,row, tree))
            idx=idx+1
        return predictions

    def make_prediction(self,x,row, tree):
        if tree.value is not None: 
            return tree.value
        feature_val = x.iloc[row,tree.feature_index]
        idx=0
        for child in tree.children:
            if feature_val == tree.threshold[idx]:
                return self.make_prediction(x,row,child)
            idx=idx+1
        return 1
    
    def get_number_of_nodes(self,tree=None):
        if tree.children is not None:
            cnt=1
            for child in tree.children:
                cnt = cnt + self.get_number_of_nodes(child)
            return cnt
        return 1
    
    def Postpruning(self,trainData,validationData):
        Y_train = trainData.iloc[:,-1]
        Y_validate = validationData.iloc[:,-1]
        training_scores = []
        validation_scores = []
        nodes=[]
        nodes.append(self.get_number_of_nodes(self.root))
        Y_train_pred = self.predict(trainData.iloc[:,:-1])
        Y_validate_pred = self.predict(validationData.iloc[:,:-1])
        initial_training_score = accuracy_score(Y_train, Y_train_pred)
        initial_validation_score = accuracy_score(Y_validate, Y_validate_pred)
        training_scores.append(initial_training_score)
        validation_scores.append(initial_validation_score)
        self.prune(self.root,trainData,validationData,training_scores,validation_scores,nodes,trainData,validationData)
        return [nodes,training_scores,validation_scores]
    
    def prune(self,tree,trainData,validationData,training_scores,validation_scores,nodes,complete_trainData,complete_validationData):
        
        if len(trainData)==0 or len(validationData)==0:
            return
        if tree.children is not None:
            idx=0
            for child in tree.children:
                child_training_dataset = self.dataset_split(trainData,tree.feature_index,tree.threshold[idx])
                child_validation_dataset = self.dataset_split(validationData,tree.feature_index,tree.threshold[idx])

                self.prune(child,child_training_dataset,child_validation_dataset,training_scores,validation_scores,nodes,complete_trainData,complete_validationData)
                idx=idx+1

        if tree.children is not None:
            if len(tree.children)>1:
                
                initial_score_train = 0
                initial_score_validate = 0
                cnt_train=0
                cnt_validate=0

                idx=0
                for child in tree.children:

                    child_validation_dataset = self.dataset_split(validationData,tree.feature_index,tree.threshold[idx])
                    if len(child_validation_dataset)==0:
                        idx+=1
                        continue
                    child_validate_y = child_validation_dataset.iloc[:,-1]

                    child_validate_ypred = self.predict(child_validation_dataset.iloc[:,:-1],child)

                    initial_score_validate = (initial_score_validate*cnt_validate + accuracy_score(child_validate_ypred,child_validate_y)*len(child_validate_ypred))/(cnt_validate+len(child_validate_ypred))

                    cnt_validate += len(child_validate_ypred)
                    idx+=1
    
                temp = tree.children
                tree.children = None
                tree.value = self.getLeafValue(trainData.iloc[:,-1])
                y_validate_pred_parent = self.predict(validationData.iloc[:,:-1],tree)
                new_accuracy_validate = accuracy_score(y_validate_pred_parent,validationData.iloc[:,-1])
                if new_accuracy_validate < initial_score_validate:
                    tree.value = None
                    tree.children=temp
                else:
                    y_train_data = complete_trainData.iloc[:,-1]
                    y_train_pred_data = self.predict(complete_trainData.iloc[:,:-1])
                    new_trainData_accuracy = accuracy_score(y_train_data,y_train_pred_data)
                    y_validate_data = complete_validationData.iloc[:,-1]
                    y_validate_pred_data = self.predict(complete_validationData.iloc[:,:-1])
                    new_validationData_accuracy = accuracy_score(y_validate_data,y_validate_pred_data)
                    if new_trainData_accuracy < new_validationData_accuracy:
                        tree.value = None
                        tree.children=temp
                        return
                    validation_scores.append(new_validationData_accuracy)
                    training_scores.append(new_trainData_accuracy)
                    nodes.append(self.get_number_of_nodes(self.root))
```

# Random Forest Classifier

```python
class RandomForestClassifier():
    def __init__(self,array_of_trees=[]):
        self.array_of_trees=array_of_trees
        
    def build_random_forest(self,trainData,validationData,testData):
        temp_store_datasets=[]
        for i in range(0,10):
            training_dataset = self.create_bootstrapped_datasets(trainData)
            tempClassifier = DecisionTreeClassifier(min_samples_split=2,maxDepth=3)
            tempClassifier.fit(training_dataset)
            self.array_of_trees.append(tempClassifier)
            temp_store_datasets.append(training_dataset)
        
        print("Under the training dataset, pre-pruning accuracy score of the forest is",self.predict(trainData))
        print("Under the validation dataset, pre-pruning accuracy score of the forest is",self.predict(validationData))
        print("Under the testing dataset, pre-pruning accuracy score of the forest is",self.predict(testData))
        for i in range(0,10):
            self.array_of_trees[i].Postpruning(temp_store_datasets[i])
        
        print("Under the training dataset, post-pruning accuracy score of the forest is",self.predict(trainData))
        print("Under the validation dataset, post-pruning accuracy score of the forest is",self.predict(validationData))
        print("Under the testing dataset, post-pruning accuracy score of the forest is",self.predict(testData))
        
    def create_bootstrapped_datasets(self,data):
        random_number = np.random.default_rng()
        sampleSize = data.shape[0]
        indices = random_number.choice(sampleSize, size = sampleSize)
        return data.iloc[indices]
    
    def fit(self, trainData,validationData,testData):
        self.build_random_forest(trainData,validationData,testData)
        
    def printAll(self):
        for i in range(len(self.array_of_trees)):
            print("-------------------- Tree",i+1,"-------------------")
            self.array_of_trees[i].print_tree()
            
    def predict(self,testing_dataset):
        idx=0
        predictions = None
        for i in range(0,10):
            X_test = testing_dataset.iloc[:,:-1]
            Y_pred = self.array_of_trees[idx].predict(X_test) 
            idx+=1
            if predictions is None:
                predictions = Y_pred
            else:
                for z in range(len(Y_pred)):
                    predictions[z] += Y_pred[z]
        size = (idx+1)/2
        for i in range(len(predictions)):
            if predictions[i]<size:
                predictions[i]=0
            else:
                predictions[i]=1
        return accuracy_score(predictions,testing_dataset.iloc[:,-1])
                
```
